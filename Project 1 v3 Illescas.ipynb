### A spin on the original project #1, where rather then spliting the fraction over and over again, the fraction is now a class. Shrinking the code further and practicing the rules and laws of classes.

```python
class Fraction:
    def __init__(self,k):
        
        #fraction making
        g = []
        g = k.split("/")
        if len(g) == 1:
            self.numerator = int(g[0])
            self.denominator = 1
        else:
            self.numerator = int(g[0])
            self.denominator = int(g[1])
        #fraction making end
        
        #reducing
        r1 = self.numerator
        r2 = self.denominator
        while r2:
            r3 = r1 % r2
            r1 = r2
            r2 = r3
        self.numerator = self.numerator // r1
        self.denominator = self.denominator // r1
        #reducing
        
        #stringing
        if self.denominator == 1:
            self.string = str(self.numerator)
            
        elif self.denominator == -1:
            self.string = str(-self.numerator)
            
        else:
            self.string = str(self.numerator) + "/" + str(self.denominator)
        #stringing

class Math:    
    
    def __init__(self,k,j):
        
        def frac_ready(k,j):
            if j == 1:
                return str(k)
            if j == -1:
                return str(-k)
            return str(k) + "/" + str(j)
        #add
        a = k.numerator*j.denominator + k.denominator*j.numerator
        b = k.denominator*j.denominator
        self.add = frac_ready(a,b)
        self.add = Fraction(self.add)
        #end add
        
        #sub
        a = k.numerator*j.denominator - k.denominator*j.numerator
        b = k.denominator*j.denominator
        self.sub = frac_ready(a,b)
        self.sub = Fraction(self.sub)  
        #end sub
        
        #mul
        a = k.numerator*j.numerator
        b = k.denominator*j.denominator
        self.mul = frac_ready(a,b)
        self.mul = Fraction(self.mul)  
        #end mul
        
        #div
        a = k.numerator*j.denominator
        b = k.denominator*j.numerator
        self.div = frac_ready(a,b)
        self.div = Fraction(self.div)  
        #end div
            

while True:
    #loop
    s = input("Press q to quit or anyting else to continue: ")
    if s == 'q':
        print("Program exits.")
        break
        
    f0 = input("enter a fraction of integers: ")
    f1 = input("enter a fraction of integers: ")
    
    f0 = Fraction(f0)
    
    f1 = Fraction(f1)
    
    f = Math(f0,f1)
    
    print(f0.string, " + ", f1.string, " = ", f.add.string)

    print(f0.string, " - ", f1.string, " = ", f.sub.string)

    print(f0.string, " * ", f1.string, " = ", f.mul.string)

    print(f0.string, " / ", f1.string, " = ", f.div.string)
```
