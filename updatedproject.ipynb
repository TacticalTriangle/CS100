```python
# gcd func
def gcd(a, b):
    r1 = a
    r2 = b
    while r2:
        r3 = r1 % r2
        r1 = r2
        r2 = r3
    return r1
# end gcd
```

```python
# spaces cause i'm lazy and it makes it easier on the eyes to see things not jumbled on top of each other
def space():
    print(" ")
```

```python
# reduce function: string to string

def reduce(f):
    f = f.split("/")
    if len(f) == 1:  # special case when fraction is a intiger
        return f[0]
    n = int(f[0])
    d = int(f[1])
    if n % d == 0:  # special case such as 6/3 returns 2
        return str(n//d)
    g = gcd(n, d)
    f = str(n//g) + "/" + str(d//g)  # // for integer division
    return f
# end reduce
```

```python
# add function: add two fractions

def add(f1, f2):
    # split f1
    f1 = f1.split("/")
    n1 = int(f1[0])
    if len(f1) == 1:
        d1 = 1
    else:
        d1 = int(f1[1])

    # split f2
    f2 = f2.split("/")
    n2 = int(f2[0])
    if len(f2) == 1:
        d2 = 1
    else:
        d2 = int(f2[1])

    # add and reduce
    d = d1*d2
    n = n1*d2+n2*d1
    f = str(n) + "/" + str(d)
    f = reduce(f)
    return f

#end add
```

```python
# sub function: subtract two fractions
    
def sub(f1, f2):
    # spliting of the frac
    f1 = f1.split("/")
    n1 = int(f1[0])
    if len(f1) == 1:
        d1 = 1
    else:
        d1 = int(f1[1])

    # spliting of the frac
    f2 = f2.split("/")
    n2 = int(f2[0])
    if len(f2) == 1:
        d2 = 1
    else:
        d2 = int(f2[1])

    # subtract and reduce
    if n1/d1 < n2/d2:
        d = d1*d2
        n = n2*d1-n1*d2
        f = str(n) + "/" + str(d)
        f = reduce(f)
    else:
        d = d1*d2
        n = n1*d2-n2*d1
        f = str(n) + "/" + str(d)
        f = reduce(f)
    return f

#end sub
```

```python
# mul function: multiply two fractions
    
def mul(f1, f2):
    # spliting of the frac
    f1 = f1.split("/")
    n1 = int(f1[0])
    if len(f1) == 1:
        d1 = 1
    else:
        d1 = int(f1[1])

    # spliting of the frac
    f2 = f2.split("/")
    n2 = int(f2[0])
    if len(f2) == 1:
        d2 = 1
    else:
        d2 = int(f2[1])

    # multiply and reduce
    d = d1*d2
    n = n1*n2
    f = str(n) + "/" + str(d)
    f = reduce(f)
    return f

#end mul
```

```python
# div function: divide two fractions
    
def div(f1, f2):
    # spliting of the frac
    f1 = f1.split("/")
    n1 = int(f1[0])
    if len(f1) == 1:
        d1 = 1
    else:
        d1 = int(f1[1])

    # spliting of the frac
    f2 = f2.split("/")
    n2 = int(f2[0])
    if len(f2) == 1:
        d2 = 1
    else:
        d2 = int(f2[1])

    # divide and reduce
    d = d1*n2
    n = n1*d2
    f = str(n) + "/" + str(d)
    f = reduce(f)
    return f

#end div
```
